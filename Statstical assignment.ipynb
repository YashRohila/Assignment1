1. Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss nominal, ordinal, interval, and ratio scales.
Data can be categorized as qualitative (categorical) or quantitative (numerical):

- Qualitative Data: Descriptive information that classifies items into groups.
Nominal Scale: Categories without order (e.g., types of fruits: apple, banana, cherry).
- Ordinal Scale: Categories with a meaningful order but no uniform difference (e.g., movie ratings: 1 star, 2 stars, 3 stars).
Quantitative Data: Numeric data that can be measured and analyzed.
- Interval Scale: Numeric data with equal intervals but no true zero (e.g., temperature in Celsius).
Ratio Scale: Numeric data with a true zero, allowing meaningful ratios (e.g., height, weight).
- 2. What are the measures of central tendency, and when should you use each? Discuss the mean, median, and mode with examples and situations where each is appropriate.
Measures of central tendency summarize a dataset with a single representative value:

Mean: The arithmetic average of all values. Best for symmetric distributions without outliers (e.g., average height of students).
Median: The middle value when data is ordered. Best for skewed distributions or when outliers are present (e.g., median household income).
Mode: The most frequently occurring value. Useful for categorical data or to identify the most common item (e.g., the most common shoe size).
- 3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?
Dispersion refers to the extent to which data points differ from the central value.

Variance: Measures the average squared deviation from the mean, providing a sense of data spread.
Standard Deviation: The square root of variance, indicating how much data typically deviates from the mean. A smaller standard deviation means the data is more concentrated around the mean.
- 4. What is a box plot, and what can it tell you about the distribution of data?
A box plot visually represents the distribution of data through five summary statistics: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It highlights:

Central tendency (median).
Spread (interquartile range).
Presence of outliers (data points outside the whiskers).
Skewness in the data distribution.
- 5. Discuss the role of random sampling in making inferences about populations.
Random sampling ensures each individual in a population has an equal chance of being selected. It:

Reduces bias, ensuring representative samples.
Allows researchers to generalize findings to the entire population.
Provides a basis for statistical inference and hypothesis testing.
- 6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?
Skewness measures the asymmetry of a dataset's distribution:

Positive Skew: The tail on the right is longer; most data values are concentrated on the left.
Negative Skew: The tail on the left is longer; most data values are concentrated on the right.
Skewness affects measures of central tendency. For example, in positively skewed data, the mean is greater than the median.

- 9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).
The normal distribution is symmetric, bell-shaped, and defined by its mean and standard deviation.

The empirical rule states:
68% of data falls within 1 standard deviation of the mean.
95% within 2 standard deviations.
99.7% within 3 standard deviations.

- 11. Explain what a random variable is and differentiate between discrete and continuous random variables.
A random variable represents a numerical outcome of a random process:

Discrete Random Variable: Takes countable values (e.g., number of heads in coin tosses).
Continuous Random Variable: Takes any value within a range (e.g., height of individuals).
